```python
from bufkit import ingest, surface
from datetime import datetime
```

```python
# Ingest model data from prior date
bufr_file = ingest('KGCN', 'GFS', datetime(2024, 1, 25, 18))
```

```python
# Get surface data from file
df = surface(bufr_file)
df
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(16, 9))
plt.plot(df.TIME, df.T2MS, linestyle='-', color='r', label='Temperature')
plt.plot(df.TIME, df.TD2M, linestyle='-', color='g', label='Dewpoint')
plt.title(f'KGCN Meteogram')
plt.xlabel('Time (UTC)')
plt.ylabel('Temperature (Â°F)')
plt.show()
```
